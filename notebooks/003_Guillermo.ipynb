```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/temixco.csv',index_col=0,parse_dates=True)
```

```python
df.plot(subplots=True,figsize=(20,14));
```

```python
df.groupby(by=([df.index.hour,df.index.minute]))['To'].mean()
```

```python
df.groupby(by=([df.index.hour,df.index.minute]))['RH'].mean()
```

```python
df.groupby(by=(df.index.strftime('%H:%M')))['To'].mean()
```

```python
df.groupby(by=([df.index.hour,df.index.minute]))['To'].mean().plot()
```

```python
df.groupby(by=(df.index.strftime('%H:%M')))['To'].mean().plot()
```
